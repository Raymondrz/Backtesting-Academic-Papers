```python
regfin=pd.read_excel('reg12.xlsx','Sheet1')
```

```python
regfin
```

```python
regfin['SpotChange']=regfin['Close T']-regfin['Close']
regfin['futchange1']=regfin['UX1 T']-regfin['UX1 Index']
regfin['futchange2']=regfin['UX2 T']-regfin['UX2 Index']
regfin['futchange3']=regfin['UX3 T']-regfin['UX3 Index']
regfin['futchange4']=regfin['UX4 T']-regfin['UX4 Index']
regfin['futchange5']=regfin['UX5 T']-regfin['UX5 Index']
regfin['futchange6']=regfin['UX6 T']-regfin['UX6 Index']
regfin['futchange7']=regfin['UX7 T']-regfin['UX7 Index']
regfin['futchange8']=regfin['UX7 T']-regfin['UX8 Index']


regfin['change1']=regfin['UX1 Index']-regfin['Close']
regfin['change2']=regfin['UX2 Index']-regfin['Close']
regfin['change3']=regfin['UX3 Index']-regfin['Close']
regfin['change4']=regfin['UX4 Index']-regfin['Close']
regfin['change5']=regfin['UX5 Index']-regfin['Close']
regfin['change6']=regfin['UX6 Index']-regfin['Close']
regfin['change7']=regfin['UX7 Index']-regfin['Close']
regfin['change8']=regfin['UX8 Index']-regfin['Close']
```

## 1st Table

```python
list2=['change1','change2','change3','change4','change5','change6','change7','change8']
intercept=[]
tstat=[]
r=[]
for nam in list2:
    model=sm.OLS(regfin['SpotChange'],sm.add_constant(regfin[nam]))
    intercept.append(model.fit().params[1])
    tstat.append(model.fit().tvalues[1])
    r.append(model.fit().rsquared_adj)
```

```python
intercept
tstat
r
```

```python
table1=pd.DataFrame(index=['UX1','UX2','UX3','UX4','UX5','UX6','UX7','UX8'],columns=['beta1','tstat','r'])
```

```python
table1['beta1']=intercept
table1['tstat']=tstat
table1['r']=r
table1=table1.T
table1.style.set_caption("Spot vix on basis : ST âˆ’St = Î±1+Î²1Â·(Ft,T âˆ’St)+ Îµ1,t")
```

```python
list2=['change1','change2','change3','change4','change5','change6','change7','change8']
list3=['futchange1','futchange2','futchange3','futchange4','futchange5','futchange6','futchange7','futchange8']

intercept=[]
tstat=[]
r=[]
for nim,nam in list(zip(list3,list2)):
    model=sm.OLS(regfin[nim],sm.add_constant(regfin[nam]))
    intercept.append(model.fit().params[1])
    tstat.append(model.fit().tvalues[1])
    r.append(model.fit().rsquared_adj)
```
